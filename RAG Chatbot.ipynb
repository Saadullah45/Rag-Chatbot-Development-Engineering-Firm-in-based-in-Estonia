```python
import os
import torch
from pinecone import Pinecone
from sentence_transformers import SentenceTransformer
from transformers import AutoTokenizer, AutoModelForCausalLM

# ✅ Hugging Face Authentication
hf_token = ""  

# ✅ Load multilingual embedding model
embedding_model = SentenceTransformer("intfloat/multilingual-e5-large-instruct")

# ✅ Initialize Pinecone
pc = Pinecone(api_key="") 
index = pc.Index("multilingual-e5-index")           

# ✅ Load Mistral 7B Instruct model (gated repo - needs token)
model_id = "mistralai/Mistral-7B-Instruct-v0.3"
tokenizer = AutoTokenizer.from_pretrained(model_id, token=hf_token)
llm = AutoModelForCausalLM.from_pretrained(
    model_id,
    torch_dtype=torch.float16,
    device_map="auto",
    token=hf_token
)

```

##  rag pipeline

```python
import re

# --- Clean non-English or unnecessary parts of query ---
def clean_query(query):
    return re.sub(r"(?i)vasta[ke]*.*?inglise keeles", "", query).strip()

# --- Retrieve relevant context chunks based on semantic similarity ---
def retrieve_context(user_query, top_k=3, similarity_threshold=0.75):
    query_embedding = embedding_model.encode(user_query).tolist()
    results = index.query(vector=query_embedding, top_k=top_k, include_metadata=True)
    context_chunks = []
    for match in results["matches"]:
        score = match.get("score", 0)
        if score >= similarity_threshold:
            text = match["metadata"].get("text", "")
            if len(text.strip()) > 50:
                context_chunks.append(text)
    return "\n".join(context_chunks)

# --- Build a clean and instructive prompt for the LLM ---
def build_prompt(context, question):
    return f"""You are a helpful assistant. Use ONLY the context provided below to answer the question. If the answer is not in the context, reply with: "I don't have enough information to answer that."

Context:
{context}

---

Question: {question}
Answer:"""

# --- Clean filler words or unnecessary model replies ---
def clean_response(text):
    return re.sub(r"^(Yes,|Sure,|Of course,)\s*", "", text, flags=re.IGNORECASE).strip()

# --- Generate the final answer using the language model ---
def generate_answer_mistral(context, query, max_new_tokens=512):
    prompt = build_prompt(context, query)
    print("🔥 Prompt sent to model:\n", prompt)  # Debug

    inputs = tokenizer(prompt, return_tensors="pt").to(llm.device)
    outputs = llm.generate(**inputs, max_new_tokens=max_new_tokens, do_sample=False, temperature=0.2)
    full_output = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Extract only the part after "Answer:"
    answer_marker = "Answer:"
    if answer_marker in full_output:
        answer = full_output.split(answer_marker, 1)[1].strip()
    else:
        answer = full_output.strip()

    return clean_response(answer)

# --- Main RAG pipeline ---
def rag_pipeline(user_query):
    cleaned_query = clean_query(user_query)
    context = retrieve_context(cleaned_query)
    print("📚 Retrieved Context:\n", context)
    response = generate_answer_mistral(context, cleaned_query)
    print("\n💬 Response:\n", response)
```
